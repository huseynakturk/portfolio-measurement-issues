## IMBD Weighted Rating


weighted_rating = (v/(v+M)*r) +(M/(v+M)*C)

r = vote average

v = vote count

M = minumum votes required to be listed in the top 250 | 250 için gerekli minumum oy değeri

C = the mean vote across the whole report(currently 7.0) | tüm filmlerin genel ortalaması

```python
# fonksiyon oluşturma 

def weighted_rating(r,v,M,C):
    return ((v /(v+M)*r) + (M/(v+M)*C))
            
```

```python
df["weighted_rating"] = weighted_rating(df["vote_average"],df["vote_count"],3500,8)
df
```

## Bayes Ortalama Derecelendirme Puanı ( BAR Score )



```python
# bayes ortalama ile hesaplanması
import math
import scipy.stats as st

def bayesian_average_rating(n,confidence=0.95):
    if sum(n) == 0:
        return 0
    K = len(n)
    z = st.norm.ppf(1-(1-confidence) / 2 )
    N = sum(n)
    first_part = 0.0
    second_part = 0.0
    
    for k,n_k in enumerate(n):
        first_part += (k+1)*(n[k]+1) / (N + K)
        second_part += (k+1) * (k+1) * (n[k] +1) / (N + K)
    score = first_part - z * math.sqrt((second_part - first_part * first_part) / (N+K+1))
    return score

#esaretin bedeli
#puan bilgileri giriyoruz
bayesian_average_rating([34733, 4355, 4704, 6561, 13515, 26183, 87368, 273082, 600260, 1295351])

```
